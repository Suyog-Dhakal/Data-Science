```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#dataset provided by seaborn
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

## Scatter Plot

```python
#hue adds color
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day")
```

## Histogram 

```python
#kde = plotting the line
#higher the bins value more information on the graph
sns.histplot(tips['tip'],kde=True,bins=25)
```

## Distribution plot

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

## Bar plot

```python
sns.barplot(x="day",y="tip", data=tips)
```

```python
sns.barplot(x="sex",y="tip", data=tips)
```

## Box Plot

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

## Strip Plot

```python
sns.stripplot(x="day",y="tip",data=tips, hue="sex", dodge=True)
```

## Joint plot 

```python
#joint plot is the combination of distribution plot and scatter plot
#reg = regression
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True)
```

```python
#hexagon
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex")
```

## Pair Plot

```python
#using titanic dataset
#pairplot parameters doesnot accpet categorical value so only choosing numeric value

sns.pairplot(titanic.select_dtypes(['number']),hue="pclass")
```

## Heat map

```python
# correlation
titanic.corr()
```

```python
#annot = true displays the values inside the box
sns.heatmap(titanic.corr(), annot=True, cmap="YlGnBu")
```
